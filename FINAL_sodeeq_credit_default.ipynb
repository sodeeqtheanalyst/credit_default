```python
# Handle outliers using the IQR method
```

```python
def treat_outliers_iqr(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1  # Calculate the interquartile range
    
    # Define the acceptable range (1.5 times the IQR)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with upper and lower bounds
    df[column] = df[column].apply(lambda x: upper_bound if x > upper_bound else x)
    df[column] = df[column].apply(lambda x: lower_bound if x < lower_bound else x)

# Apply the outlier treatment to the necessary columns
# List of numerical columns where outliers are to be treated
numerical_columns = ['_Cash_Flow_Per_Share', '_Total_debt_to_Total_net_worth', '_Cash_to_Total_Assets', '_Current_Liability_to_Current_Assets']

for column in numerical_columns:
    treat_outliers_iqr(df, column)

# Now the outliers in the specified columns are treated
```

```python
columns_for_analysis = ['_Cash_Flow_Per_Share', '_Total_debt_to_Total_net_worth', '_Cash_to_Total_Assets', 
                        '_Current_Liability_to_Current_Assets', 'Default']

# Univariate Analysis with charts arranged in a grid
def univariate_analysis(df, columns):
    fig, axes = plt.subplots(2, 2, figsize=(18, 10))  
    axes = axes.flatten() 
    
    for i, column in enumerate(columns):
        sns.histplot(df[column], kde=True, bins=30, color='blue', ax=axes[i])
        axes[i].set_title(f'Distribution of {column}')
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('Frequency')
        print(f"Summary statistics for {column}:\n", df[column].describe())
        print("\n" + "="*50 + "\n")
    
    # Hide the last unused subplot (if you have less than 6 charts)
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Adjust subplots to fit into figure area
    plt.show()

# Perform Univariate Analysis on selected columns
univariate_analysis(df, columns_for_analysis[:-1])  # excluding 'Default' as it's a target variable
```

```python
##Bivariate Analysis
```

```python
def bivariate_analysis(df, columns, target):
    fig, axes = plt.subplots(2, 2, figsize=(18, 10)) 
    axes = axes.flatten()  
    
    for i, column in enumerate(columns):
        sns.boxplot(x=target, y=column, data=df, ax=axes[i])
        axes[i].set_title(f'{column} vs {target}')
        axes[i].set_xlabel(target)
        axes[i].set_ylabel(column)

        correlation = df[[column, target]].corr().iloc[0, 1]
        print(f"Correlation between {column} and {target}: {correlation}")
        print("\n" + "="*50 + "\n")

    # Hide the last unused subplot (if you have less than 6 charts)
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Adjust subplots to fit into figure area
    plt.show()

# Perform Bivariate Analysis between selected columns and the 'Default' target
bivariate_analysis(df, columns_for_analysis[:-1], 'Default')
```

##Train-Test Split

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Select relevant columns for analysis
columns_for_analysis = ['_Cash_Flow_Per_Share', '_Total_debt_to_Total_net_worth', '_Cash_to_Total_Assets', 
                        '_Current_Liability_to_Current_Assets', 'Default']

# Train-Test Split
X = df[columns_for_analysis[:-1]]  # Features (independent variables)
y = df['Default']  # Target (dependent variable)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Define a function for univariate and bivariate analysis
def plot_analysis(df, columns, target=None, kind='univariate'):
    fig, axes = plt.subplots(2, 2, figsize=(18, 10))
    axes = axes.flatten()
    
    for i, column in enumerate(columns):
        if kind == 'univariate':
            sns.histplot(df[column], kde=True, bins=30, color='blue', ax=axes[i])
            axes[i].set_title(f'Distribution of {column}')
        elif kind == 'bivariate':
            sns.boxplot(x=target, y=column, data=df, ax=axes[i])
            axes[i].set_title(f'{column} vs {target}')
        
        axes[i].set_xlabel(column)
    
    for j in range(i+1, len(axes)):  # Hide unused subplots
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

# Univariate Analysis on training data
plot_analysis(X_train, columns_for_analysis[:-1], kind='univariate')

# Bivariate Analysis with the target 'Default'
plot_analysis(df, columns_for_analysis[:-1], target='Default', kind='bivariate')
```

```python
#LOGISTIC REGRESSION MODEL
```

```python
import statsmodels.api as sm
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np


# Select relevant columns (assumed important variables for the model based on your analysis)
important_columns = ['_Cash_Flow_Per_Share', '_Total_debt_to_Total_net_worth', '_Cash_to_Total_Assets', 
                     '_Current_Liability_to_Current_Assets', 'Default']

# Split the data into features (X) and target (y)
X = df[important_columns[:-1]]  # Independent variables
y = df['Default']  # Dependent variable

# Train-Test Split (67% train, 33% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Add a constant term to the independent variables (for the intercept)
X_train_sm = sm.add_constant(X_train)

# Build Logistic Regression Model
logit_model = sm.Logit(y_train, X_train_sm)
result = logit_model.fit()

# Print the summary of the logistic regression model
print(result.summary())

# Predict probabilities for the test set
X_test_sm = sm.add_constant(X_test)
y_pred_prob = result.predict(X_test_sm)

# Evaluate the model: ROC curve to choose an optimum cutoff
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Dashed diagonal line for random guess
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression')
plt.legend(loc="lower right")
plt.show()

# Optimum cutoff based on ROC curve
# We can choose the threshold where the True Positive Rate (TPR) is high and False Positive Rate (FPR) is low.
optimum_cutoff_index = np.argmax(tpr - fpr)  # Maximizing TPR - FPR gives best cutoff
optimum_cutoff = thresholds[optimum_cutoff_index]
print(f'Optimum cutoff: {optimum_cutoff:.2f}')

# Apply the optimum cutoff to make final predictions
y_pred_final = np.where(y_pred_prob >= optimum_cutoff, 1, 0)

# Print the final predicted values and compare with true values
print("Predicted classes with optimum cutoff:", y_pred_final)
print("True classes:", y_test.values)
```

VALIDATATION ON TEST DATASET

```python
# Import necessary libraries for performance metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Predict probabilities for the test set
X_test_sm = sm.add_constant(X_test)
y_pred_prob = result.predict(X_test_sm)

# Apply the optimum cutoff (previously calculated)
y_pred_final = np.where(y_pred_prob >= optimum_cutoff, 1, 0)

# Performance Metrics
accuracy = accuracy_score(y_test, y_pred_final)
conf_matrix = confusion_matrix(y_test, y_pred_final)
class_report = classification_report(y_test, y_pred_final)

# Print the performance metrics
print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```
